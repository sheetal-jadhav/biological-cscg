```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from hmmlearn import hmm
```

# Step 1: define markov chain states and transition matrix
* define all posible chain states
* sepcify transition probabilities 
* use this to later simulate markov chain

```python
states = ["sunny","rainy"]

transition_matrix = np.array([[0.8,0.2],
                             [0.4,0.6]])

print("\nTransition Matrix")
print(transition_matrix)

```

# Step 2: Simulate Markov Chain Sequence
* generate sequence of states from transition matrix
* state from an initial state
* randomly choose next states according to trannsition probabilities

```python
def simulate_markov_chain(start_state, n_steps=20):
    current_state = states.index(start_state)
    sequence = [start_state]
    for _ in range(n_steps):
        current_state = np.random.choice([0,1], p=transition_matrix[current_state])
        sequence.append(states[current_state])
    return sequence

mc_sequence = simulate_markov_chain('sunny',15)
print("\n Generated Markov Chain Sequence:")
print(mc_sequence)
```

# Step 3: Define HMM paramterers
* define hidden states and observations for HMM
* set initial state probabilities, transition matrix and emission matrix
* these paramters define the behavior of HMM

```python
hidden_states = ["Hot", "Cold"]
observations = ["IceCream", "Coffee"]

model = hmm.MultinomialHMM(n_components=2, n_iter=50)
model.transmat_ = np.array([
    [0.7, 0.3],  
    [0.4, 0.6]    
])
model.emissionprob_ = np.array([
    [0.8, 0.2],  
    [0.3, 0.7]   
])
model.startprob_ = np.array([0.6, 0.4])

print("\nTransition Matrix (Hidden States):")
print(model.transmat_)
print("\nEmission Matrix:")
print(model.emissionprob_)
```
